```python
import seaborn as sns
%matplotlib inline
```

```python
plan = sns.load_dataset('planets')
```

```python
plan.head()
```

```python
sns.distplot(plan['mass'])
```

```python
sns.distplot(plan['mass'], bins = 50)
```

```python
sns.jointplot(x='orbital_period', y='mass', data=plan)
```

```python
sns.jointplot(x='mass', y='year',data=plan,kind='kde')
```

```python
sns.jointplot(x='mass', y='year',data=plan,kind='hex')
```

```python
sns.pairplot(plan)
```

```python
sns.barplot(x='year',y='mass',data=plan)
```

```python
import matplotlib.pyplot as plt
g = sns.FacetGrid(plan, col="mass", row="orbital_period")
g = g.map(plt.hist,'year')
```
